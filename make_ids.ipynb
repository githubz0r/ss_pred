```python
import itertools
import functools
import time
test='eeeehhhhhhhhcccceeeeehhhcccccceeeeeeeeeeehhcccchhhhhhhhhh'
```

```python
splits = test.split('(?<=(.))(?!\\1)')
```

```python
itertools.groupby(test)
```

```python
splits = ["".join(grp) for val, grp in itertools.groupby(test)]
```

```python
sum([len(i) for i in splits])
```

```python
len(test)
```

```python
functools.reduce(lambda x,y:x+y, splits)
```

```python
def return_grps(string):
    cur_symbol = string[0]
    grps = []
    subgrp = []
    for i in string:
        if i == cur_symbol:
            subgrp.append(i)
        else:
            grps.append("".join(subgrp))
            cur_symbol = i
            subgrp = [i]
    grps.append("".join(subgrp)) # last group
    return grps

return_grps(test)
```
